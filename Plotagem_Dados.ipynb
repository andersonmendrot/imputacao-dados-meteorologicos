# Análise Comparativa de Técnicas de Aprendizado para Imputação de Dados Faltantes em Séries Temporais de Dados Meteorológicos

O código abaixo apresenta um exemplo de plotagem dos dados meteorológicos em um período de cinco anos (os dados incluem os anos de 1961 a 2020). O exemplo utiliza as bibliotecas pandas, matplotlib e pyplot. Pela plotagem, é possivel perceber a necessidade do projeto realizado, pois a partir dele são testadas as técnicas de MLP e Regressão Linear para imputação de dados faltantes. Será criado outro notebook pada demonstração dos resultados obtidos.

```python
#inicialmente, devem ser importadas as bibliotecas a serem utilizadas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
```

```python
#o dataset é carregado com o comando abaixo
df = pd.read_csv("https://raw.githubusercontent.com/andersonmendrot/imputacao-dados-meteorologicos/master/dataset_mirante_santana.txt",
                    delimiter = ';',
                    skipinitialspace=True,
                    usecols=["Data", "Hora", "TempMaxima"],
                    skiprows=16) 
```

```python
#por conta de as informações de data estarem em formato de texto, é necessário que seja transformada para formato de data
df['Data'] = pd.to_datetime(df['Data'])
```

```python
#o dataset apresenta medicoes de temperatura maxima à meia noite, e temperatura minima ao meio dia, e serão consideradas apenas as medições à meia noite
df = df[df.Hora == 0] 
```

```python
#a coluna de data será nosso indice para plotagem
df.set_index('Data', inplace=True)
```

```python
#os dados serão selecionados para plotagem em 3 anos consecutivos, de 1980 a 1982
df_dois_anos = df.loc['1980-01-01':'1982-01-01']
```

```python
#é realizada então a plotagem dos dados utilizando o parâmetro de temperatura máxima no eixo y
df_dois_anos.plot(y='TempMaxima')
```
